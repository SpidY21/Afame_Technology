```python
col_to_drop = ['trans_date_trans_time', 'cc_num', 'merchant', 'first', 'last', 'street', 'city', 'state', 'job', 'dob', 'trans_num']
```

```python
df = df_train.drop(col_to_drop,axis=1)
```

```python
df.gender = df.gender.replace('M', 1).replace('F', 0)
```

```python
df.head()
```

```python
df.info()
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df = df.replace(False, 0).replace(True, 1)
```

```python
df.shape
```

# Data Spliting

```python
X = df.drop('is_fraud',axis=1)
y = df['is_fraud']
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

# VIF Analysis

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['fearure'] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
vif['VIF'] = vif['VIF']
vif = vif.sort_values(by='VIF',ascending=False)
vif
```

```python
X = X.drop('long',axis=1)
```

```python
vif = pd.DataFrame()
vif['fearure'] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
vif['VIF'] = vif['VIF']
vif = vif.sort_values(by='VIF',ascending=False)
vif
```

```python
X = X.drop('unix_time',axis=1)
```

```python
vif = pd.DataFrame()
vif['fearure'] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
vif['VIF'] = vif['VIF']
vif = vif.sort_values(by='VIF',ascending=False)
vif
```

```python
X = X.drop('lat', axis=1)
vif = pd.DataFrame()
vif['fearure'] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
vif['VIF'] = vif['VIF']
vif = vif.sort_values(by='VIF',ascending=False)
vif
```

```python
X = X.drop('merch_long', axis=1)
vif = pd.DataFrame()
vif['fearure'] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
vif['VIF'] = vif['VIF']
vif = vif.sort_values(by='VIF',ascending=False)
vif
```

```python
rus = RandomUnderSampler(random_state=21)
```

```python
X_new, y_new = rus.fit_resample(X, y)
```

```python
df.head()
```

```python
df.columns, len(df.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new,test_size=0.25,random_state=21)
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaling = StandardScaler()
```

```python
X_train_scaled = scaling.fit_transform(X_train)
X_test_scaled = scaling.transform(X_test)
```

```python
X_train_scaled.shape, X_test_scaled.shape
```

# Modeling

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state=21)
lr.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, auc, roc_auc_score, roc_curve, ConfusionMatrixDisplay, confusion_matrix
```

```python
print(f'Train accuracy of the model is: {round(accuracy_score(y_train, lr.predict(X_train_scaled))*100,2)} %')
print(f'Train precision of the model is: {round(precision_score(y_train, lr.predict(X_train_scaled))*100,2)} %')
print(f'Train recall of the model is: {round(recall_score(y_train, lr.predict(X_train_scaled))*100,2)} %')
print('-'*40)
print(f'Test accuracy of the model is: {round(accuracy_score(y_test, lr.predict(X_test_scaled))*100,2)} %')
print(f'Test precision of the model is: {round(precision_score(y_test, lr.predict(X_test_scaled))*100,2)} %')
print(f'Test recall of the model is: {round(recall_score(y_test, lr.predict(X_test_scaled))*100,2)} %')
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier(random_state=21)
gbc.fit(X_train_scaled,y_train)

print(f'Train accuracy of the model is: {round(accuracy_score(y_train, gbc.predict(X_train_scaled))*100,2)} %')
print(f'Train precision of the model is: {round(precision_score(y_train, gbc.predict(X_train_scaled))*100,2)} %')
print(f'Train recall of the model is: {round(recall_score(y_train, gbc.predict(X_train_scaled))*100,2)} %')
print('-'*40)
print(f'Test accuracy of the model is: {round(accuracy_score(y_test, gbc.predict(X_test_scaled))*100,2)} %')
print(f'Test precision of the model is: {round(precision_score(y_test, gbc.predict(X_test_scaled))*100,2)} %')
print(f'Test recall of the model is: {round(recall_score(y_test, gbc.predict(X_test_scaled))*100,2)} %')
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(random_state=21)
rfc.fit(X_train_scaled,y_train)

print(f'Train accuracy of the model is: {round(accuracy_score(y_train, rfc.predict(X_train_scaled))*100,2)} %')
print(f'Train precision of the model is: {round(precision_score(y_train, rfc.predict(X_train_scaled))*100,2)} %')
print(f'Train recall of the model is: {round(recall_score(y_train, rfc.predict(X_train_scaled))*100,2)} %')
print('-'*40)
print(f'Test accuracy of the model is: {round(accuracy_score(y_test, rfc.predict(X_test_scaled))*100,2)} %')
print(f'Test precision of the model is: {round(precision_score(y_test, rfc.predict(X_test_scaled))*100,2)} %')
print(f'Test recall of the model is: {round(recall_score(y_test, rfc.predict(X_test_scaled))*100,2)} %')
```
